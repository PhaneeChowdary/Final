```python
def txtTocsv(fileName):
    import pandas as pd
    import re
    path = 'authors/' + fileName + '.txt'
    words = dict()
    with open(path, 'r') as file:
        lines = [line.strip().lower() for line in file.readlines() if line.strip()]
        pattern = ':;{}[]\\|+=-_<>,.?/!@#$%^&*()\'\"'
        for line in lines:
            for word in line.split():
                word = ''.join(filter(lambda k: k not in pattern, word))
                word = word.lstrip('“”’')
                word = word.rstrip('“”’')
                if word in words: words[word] += 1
                else: words[word] = 1
        word = words.keys()
        count = words.values()
        print('Unique words : ', len(word))
        data = {'Words': word, 'Count': count}
        df = pd.DataFrame.from_dict(data)
        filename = fileName + '.csv'
        df.to_csv(filename, index=False)
        
for i in range(1, 6):
    filename = 'a' + str(i)
    txtTocsv(filename)
    filename = 'b' + str(i)
    txtTocsv(filename)
txtTocsv('ua')
txtTocsv('ub')
```

```python
def get_dataset(author, num):
    path = author + str(num) + '.csv'
    df = pd.read_csv(path, index_col=0)
    total = df.Count.sum()
    df.loc[:, 'Count'] = df.Count/total
    df = df[df["Count"] > 0.0005]
    df.rename(columns={"Count": author+str(num)}, inplace=True)
    return df.transpose()
```

```python
df = pd.concat([ get_dataset(author, i) for author in ["a", "b"] for i in range(1, 6) ])
df = pd.concat( [df] + [ get_dataset(author, i) for author in ["u"] for i in ['a', 'b'] ])
df
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_word_1d(w1):
    plt.scatter(df.loc["a1":"a5", w1], 5 * [0], c="r")  # austen
    plt.scatter(df.loc["b1":"b5", w1], 5 * [0], c="b")  # dickens
    plt.scatter(df.loc["ua", w1], [0], c="orange")  # unknown work 1
    plt.scatter(df.loc["ub", w1], [0], c="g")  # unknown work 1
    plt.xlabel(f"{w1}")
    plt.show()
plot_word_1d("good")
```

```python
def plot_word_2d(w1, w2):
    plt.scatter(df.loc["a1":"a5", w1], df.loc["a1":"a5", w2], c="r")  # austen
    plt.scatter(df.loc["b1":"b5", w1], df.loc["b1":"b5", w2], c="b")  # dickens
    plt.scatter(df.loc["ua", w1], df.loc["ua", w2], c="orange")  # unknown work 1
    plt.scatter(df.loc["ub", w1], df.loc["ub", w2], c="g")  # unknown work 1
    plt.xlabel(f"{w1}")
    plt.ylabel(f"{w2}")
    plt.show()
plot_word_2d("good", "first")
```

```python


def plot_word_3d(w1, w2, w3):
    fig = plt.figure()
    ax = fig.add_subplot(projection="3d")
    
    ax.scatter(df.loc["a1":"a5", w1], df.loc["a1":"a5", w2], df.loc["a1":"a5", w3], c="r")  # a
    ax.scatter(df.loc["b1":"b5", w1], df.loc["b1":"b5", w2], df.loc["b1":"b5", w3], c="b")  # b
    ax.scatter(df.loc["ua", w1], df.loc["ua", w2], df.loc["ua", w3], c="orange")  # unknown work 1
    ax.scatter(df.loc["ub", w1], df.loc["ub", w2], df.loc["ub", w3], c="g")  # unknown work 1
    ax.set_xlabel(f"{w1}")
    ax.set_ylabel(f"{w2}")
    ax.set_zlabel(f"{w3}")
    plt.show()
    
plot_word_3d("of", "the", "i")
```

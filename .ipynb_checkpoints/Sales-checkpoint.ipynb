# Sales Analysis

# Definitions

- **Total Revenue** - Total revenue in sales including the tax revenue. This is the total amount of money coming in including tax and without subtracting costs. This amounts to the sum of each sales transaction expressed as `Price * Quantity + State Tax`. **Note** `State Tax` is `(Price * Quantity) * State Tax Rate`, so combining the two expressions, `Total Revenue` could be described as `Price * Quantity * (1 + State Tax Rate)`

- **Gross Profit** - Total profit in sales after subtracting tax revenue and subtracting product costs. This amounts to the sum of each sales transaction (`(Price - Cost) * Quantity)`).

- **Net Profit** - Total profit in sales after subtracting additional business expenses. This is the total amount of money left after subtracting all costs. This amounts to the expression `Gross Profit for a Given Period - All Additional Business Expenses for a Given Period`.

```python
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)
import matplotlib.pyplot as plt
%matplotlib inline
from collections import defaultdict
```

```python
customers_df = pd.read_csv('https://cs.indstate.edu/~lmay1/assets/sales-data/customers.csv')
expenses_df = pd.read_csv('https://cs.indstate.edu/~lmay1/assets/sales-data/expenses.csv')
products_df = pd.read_csv('https://cs.indstate.edu/~lmay1/assets/sales-data/products.csv')
tax_df = pd.read_csv('https://cs.indstate.edu/~lmay1/assets/sales-data/tax.csv')
customers = customers_df.copy()
```

```python
us_states = {'Alabama': 'AL', 'Alaska': 'AK', 'Arizona': 'AZ', 'Arkansas': 'AR', 'California': 'CA',
             'Colorado': 'CO', 'Connecticut': 'CT', 'Delaware': 'DE', 'District of Columbia': 'DC',
             'Florida': 'FL', 'Georgia': 'GA', 'Hawaii': 'HI', 'Idaho': 'ID', 'Illinois': 'IL',
             'Indiana': 'IN', 'Iowa': 'IA', 'Kansas': 'KS', 'Kentucky': 'KY', 'Louisiana': 'LA',
             'Maine': 'ME', 'Maryland': 'MD', 'Massachusetts': 'MA', 'Michigan': 'MI', 'Minnesota': 'MN',
             'Mississippi': 'MS', 'Missouri': 'MO', 'Montana': 'MT', 'Nebraska': 'NE', 'Nevada': 'NV',
             'New Hampshire': 'NH', 'New Jersey': 'NJ', 'New Mexico': 'NM', 'New York': 'NY', 
             'North Carolina': 'NC', 'North Dakota': 'ND', 'Ohio': 'OH', 'Oklahoma': 'OK', 'Oregon': 'OR',
             'Pennsylvania': 'PA', 'Rhode Island': 'RI', 'South Carolina': 'SC', 'South Dakota': 'SD',
             'Tennessee': 'TN', 'Texas': 'TX', 'Utah': 'UT', 'Vermont': 'VT', 'Virginia': 'VA',
             'Washington': 'WA', 'West Virginia': 'WV', 'Wisconsin': 'WI', 'Wyoming': 'WY'
}
```

```python
# Swap the switched Zip and States.

switched_cnt = 0
for i in range(len(customers)):
    x = customers['Zip'][i]
    try:
        if int(x): pass
    except:
        if x is not np.nan:
            switched_cnt += 1
            customers['Zip'][i], customers['State'][i] = customers['State'][i], customers['Zip'][i]
```

```python
features = []
percentages = []
for column in customers.columns:
    blanks = round((customers[column].isnull().sum()/len(customers))*100, 2)
    if blanks > 0:
        features.append(column)
        percentages.append(blanks)
        
features.append('Zip_State_Switched')
percentages.append(round((switched_cnt/len(customers))*100, 2))
```

```python
fig = plt.figure(figsize = (10, 5))
colors = ['green', 'red', 'orange', 'maroon', 'blue']

for feature, percent, color in zip(features, percentages, colors):
    plt.bar(feature, percent, label = feature, color = color)
    
plt.xlabel("Features")
plt.ylabel("Percentage of fields left blank")
plt.title("Bar Chart")
plt.legend()
```

```python
# Products Cost, Price

prod_price = dict()
prod_cost = dict()
for i in range(len(products_df)):
    prod = products_df.ProductId[i]
    prod_price[prod] = products_df.Price[i]
    prod_cost[prod] = products_df.Cost[i]
```

```python
# State tax rate

state_tax_rates = dict()
for i in range(len(tax_df)):
    rate = tax_df.Rate[i]
    if rate is not np.nan:
        state_tax_rates[us_states[tax_df.State[i]]] = float(tax_df.Rate[i][:-1])
```

```python
# Add state tax rates to customers

StateTaxRate = [None]*len(customers)
customers['StateTaxRate'] = StateTaxRate
for i in range(len(customers)):
    if customers.State[i] is not np.nan:
        customers['StateTaxRate'][i] = state_tax_rates[customers.State[i]]

customers.StateTaxRate.fillna(0, inplace=True)
```

```python
# Function to calculate monthly gross profit and net profit of products
def monthly_gross_net_rev():
    gross_profit = defaultdict(list)
    net_profit = defaultdict(list)
    revenue = defaultdict(list)

    #Customer tax rates
    customer_tax_rates = dict()
    for i in range(len(customers)): customer_tax_rates[customers.CustomerId[i]] = customers.StateTaxRate[i]

    for i in range(1, 13):
        if i<10: sales=pd.read_csv('https://cs.indstate.edu/~lmay1/assets/sales-data/sales-0'+ str(i) + '.csv')
        else: sales=pd.read_csv('https://cs.indstate.edu/~lmay1/assets/sales-data/sales-'+ str(i) + '.csv')

        #Find quantity of each product in this month 
        prod_qnty = dict()
        for prod in products_df.ProductId.unique():
            qnty = sales[sales.ProductId == prod].Quantity.sum()        
            prod_qnty[prod] = qnty

        #Monthly gross and net profits
        for prod in products_df.ProductId:
            gross_profit[prod].append((prod_price[prod] - prod_cost[prod]) * prod_qnty[prod])
        for prod in products_df.ProductId:
            net_profit[prod].append(gross_profit[prod][i-1] - expenses_df.Expenses[i-1])

        #Monthly revenue
        month_revenue = dict()
        df = sales.groupby(by = ['CustomerId'])
        custIds = sorted(sales.CustomerId.unique())
        for custId in custIds:
            prods = dict()
            df2 = df.get_group(custId)
            for i in range(len(df2)):
                prod = df2.ProductId[df2.index[i]]
                if prod not in prods: prods[prod] = df2.Quantity[df2.index[i]]
                else: prods[prod] += df2.Quantity[df2.index[i]]

            for prod in prods.keys():
                if prod not in month_revenue: month_revenue[prod] = prod_price[prod]*prods[prod]*(1+customer_tax_rates[custId]/100)
                else: month_revenue[prod] += prod_price[prod]*prods[prod]*(1+customer_tax_rates[custId]/100)

        for prod in month_revenue.keys():
            if prod not in revenue: revenue[prod] = [month_revenue[prod]]
            else: revenue[prod].append(month_revenue[prod])

    return gross_profit, net_profit, revenue

gross_profit, net_profit, revenue = monthly_gross_net_rev()
```

```python
# Revenue, Gross and Net profits monthly wise

month_rev = []
month_gross = []
month_net = []
prods = list(products_df.ProductId)
for i in range(12):
    month_rev.append(sum([revenue[prod][i] for prod in prods]))
    month_gross.append(sum([gross_profit[prod][i] for prod in prods]))
    month_net.append(month_gross[i] - expenses_df['Expenses'][i])
```

```python
fig = plt.figure(figsize = (10, 5))
plt.xlabel('Month')
plt.ylabel('Dollars')
plt.xticks(np.arange(0, 12, step=1), labels = ['Jan', 'Feb', 'March', 'April', 'May',  'June',
                                                'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.title('Product sales')
plt.plot(month_rev, label='Total Revenue', linewidth=4)
plt.plot(month_gross, label='Gross Profit', linewidth=3)
plt.plot(month_net, label='Net Profit', linewidth=3)

# Plot min and max net profits
# Min
min_net = float('%.2f' %min(month_net))
plt.plot(1, min_net, 'bo', label = min_net)

# Max
max_net = float('%.2f' %max(month_net))
plt.plot(9, max_net, 'ro', label=max_net)

plt.legend()
plt.show()
```

```python
def annual_gross_revenue(annual_gross_profit, revenue):
    annual_revenue = dict()
    for prod in products_df.ProductId:

        #Annual gross profit
        gross = sum(gross_profit[prod])
        annual_gross_profit[prod] = gross

        #Annual revenue
        rev = sum(revenue[prod])
        annual_revenue[prod] = rev

    return annual_gross_profit, annual_revenue

annual_gross_profit = gross_profit.copy()
annual_gross_profit, annual_revenue = annual_gross_revenue(annual_gross_profit, revenue)
```

```python
fig = plt.figure(figsize = (12, 6))
x_keys = annual_revenue.keys()
r = np.arange(len(x_keys))
width = 0.4
plt.bar(r, annual_revenue.values(), width = width, label='Annual revenue')
plt.bar(r + width, annual_gross_profit.values(), width = width, label='Annual Gross')

# Plot best and worst selling products
best = float('-inf')
worst = float('inf')
for prod in annual_revenue:
    if annual_revenue[prod] < worst:
        worst = annual_revenue[prod]
        worst_item = prod
    elif annual_revenue[prod] > best:
        best = annual_revenue[prod]
        best_item = prod
        
best_gross = float('-inf')
worst_gross = float('inf')
for prod in annual_gross_profit:
    if annual_gross_profit[prod] < worst_gross:
        worst_gross = annual_gross_profit[prod]
        worst_gross_item = prod
    elif annual_gross_profit[prod] > best_gross:
        best_gross = annual_gross_profit[prod]
        best_gross_item = prod


best = float('%.2f' %best)
worst = float('%.2f' %worst)
plt.plot(1, best, 'go', label=str(best)+best_item)
plt.plot(6, worst, 'ro', label=str(worst)+worst_item)
best_gross = float('%.2f' %best_gross)
worst_gross = float('%.2f' %worst_gross)
plt.plot(1.5, best_gross, 'yo', label=str(best_gross)+best_gross_item)
plt.plot(2.5, worst_gross, 'bo', label=str(worst_gross)+worst_gross_item)

plt.xlabel('Products')
plt.ylabel('Annual Income ($)')
plt.xticks(r + width/2, x_keys)
plt.legend([str(best_gross)+'-'+best_gross_item, str(worst_gross)+'-'+worst_gross_item, str(best)+'-'+best_item, str(worst)+'-'+worst_item, 'Annual Revenue', 'Annual Gross'])
plt.show()
```

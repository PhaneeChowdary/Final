# Make a Function

```python
def load(author, i):
    artistic_work = f"{author}{i:0>2}"
    df = pd.read_csv(f"{artistic_work}.csv", header=None, names=["Word", "Count"], index_col=0)
    total = df["Count"].sum()
    df.loc[:, "Count"] = df["Count"] / total # Column "Count" is now a "Frequency"
    df = df[df["Count"] > 0.0005]
    df.rename(columns={"Count": artistic_work}, inplace=True)
    df1 = df.transpose()
    return df1
load("austen", 1)
```

```python
df = pd.concat([ load(author, i) for author in ["austen", "dickens"] for i in range(1, 6) ])
df = pd.concat( [df] + [ load(author, i) for author in ["unknown"] for i in range(1, 3) ])
df
```

# Plot Functions

```python
# %matplotlib notebook
import matplotlib.pyplot as plt
```

## One Dimension

```python
def plot_word_1d(w1):
    plt.scatter(df.loc["austen01":"austen05", w1], 5 * [0], c="r")  # austen
    plt.scatter(df.loc["dickens01":"dickens05", w1], 5 * [0], c="b")  # dickens
    plt.scatter(df.loc["unknown01", w1], [0], c="orange")  # unknown work 1
    plt.scatter(df.loc["unknown02", w1], [0], c="g")  # unknown work 1
    plt.xlabel(f"{w1}")
    plt.show()
plot_word_1d("should")
```

## Two Dimensions

```python
def plot_word_2d(w1, w2):
    plt.scatter(df.loc["austen01":"austen05", w1], df.loc["austen01":"austen05", w2], c="r")  # austen
    plt.scatter(df.loc["dickens01":"dickens05", w1], df.loc["dickens01":"dickens05", w2], c="b")  # dickens
    plt.scatter(df.loc["unknown01", w1], df.loc["unknown01", w2], c="orange")  # unknown work 1
    plt.scatter(df.loc["unknown02", w1], df.loc["unknown02", w2], c="g")  # unknown work 1
    plt.xlabel(f"{w1}")
    plt.ylabel(f"{w2}")
    plt.show()
plot_word_2d("could", "should")
```

## Three Dimensions

```python


def plot_word_3d(w1, w2, w3):
    fig = plt.figure()
    ax = fig.add_subplot(projection="3d")
    
    ax.scatter(df.loc["austen01":"austen05", w1], df.loc["austen01":"austen05", w2], df.loc["austen01":"austen05", w3], c="r")  # austen
    ax.scatter(df.loc["dickens01":"dickens05", w1], df.loc["dickens01":"dickens05", w2], df.loc["dickens01":"dickens05", w3], c="b")  # dickens
    ax.scatter(df.loc["unknown01", w1], df.loc["unknown01", w2], df.loc["unknown01", w3], c="orange")  # unknown work 1
    ax.scatter(df.loc["unknown02", w1], df.loc["unknown02", w2], df.loc["unknown02", w3], c="g")  # unknown work 1
    ax.set_xlabel(f"{w1}")
    ax.set_ylabel(f"{w2}")
    ax.set_zlabel(f"{w3}")
    plt.show()
plot_word_3d("the", "and", "but")
```

```python
votes = [ work[:-2] for work in u1_knn3["Work"].values.tolist() ] 
votes.append("dickens") 
votes.append("dickens")
votes
```

```python
results = { candidate:votes.count(candidate) for candidate in list(set(votes)) }
results
```
